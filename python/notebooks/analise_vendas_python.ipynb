```python
# 📦 Importando as bibliotecas que vamos usar
import pandas as pd  # Para trabalhar com os dados
import seaborn as sns  # Para criar gráficos bonitos
import matplotlib.pyplot as plt  # Para personalizar os gráficos

# 🎨 Configura o estilo dos gráficos
sns.set(style="whitegrid")
```

```python
# 📂 Lendo os dados do arquivo CSV
df = pd.read_csv("dados_vendas.csv", sep=";", encoding="latin1")

# 🔍 Dando uma olhada nas primeiras linhas do nosso dataset
df.head()
```

```python
# 🔢 Agrupando os produtos e somando as quantidades vendidas
mais_vendidos = df.groupby('Produto')['Quantidade'].sum().sort_values(ascending=False)

# 📝 Mostrando a lista dos produtos mais vendidos
print("📊 Produtos mais vendidos (em quantidade):")
print(mais_vendidos)
```

```python
# 📊 Criando um gráfico de barras para os produtos mais vendidos
plt.figure(figsize=(10, 5))  # Ajusta o tamanho do gráfico
sns.barplot(x=mais_vendidos.index, y=mais_vendidos.values, palette='coolwarm')

# 🎯 Personalizando o gráfico
plt.title('Produtos Mais Vendidos')
plt.ylabel('Quantidade Vendida')
plt.xlabel('Produto')
plt.xticks(rotation=45)  # Gira os nomes para melhor leitura

# 💾 Salva o gráfico
plt.savefig("C:/Users/Aline.Aline/Desktop/GitHub/Projeto 1 - Dashboard de Vendas/analise_python/img/produtos_mais_vendidos.png") 

# 👀 Exibe o gráfico
plt.tight_layout()
plt.show()
```

## 💡 Insights: Produtos Mais Vendidos

🎧 **Fone de ouvido** foi o produto mais vendido, com 4 unidades — alta demanda por acessórios de áudio.  
🖱️ **Mouse** vem em segundo, com 3 unidades vendidas — também mostra o interesse em acessórios.  
💻 Produtos de maior valor como **Notebook, Teclado, Monitor e Smartphone** tiveram menor volume de vendas.  
📦 Pode ser interessante focar mais em itens de alta saída no estoque e reavaliar a demanda dos demais.  
📊 **Estratégias de promoção** para produtos com menos saída podem ajudar a equilibrar o mix de vendas.



```python
# 🌎 Agrupando as vendas por região
regiao_vendas = df.groupby('Região')['Valor Total'].sum().sort_values(ascending=False)

# 📊 Mostrando os valores totais por região
print("📍 Total de vendas por região:")
print(regiao_vendas)
```

```python
# 🗺️ Criando gráfico de vendas por região
plt.figure(figsize=(8, 5))
sns.barplot(x=regiao_vendas.index, y=regiao_vendas.values, palette='magma')

# 🎯 Personalizando
plt.title('Total de Vendas por Região')
plt.ylabel('Valor Total (R$)')
plt.xlabel('Região')

# 💾 Salva o gráfico
plt.savefig("C:/Users/Aline.Aline/Desktop/GitHub/Projeto 1 - Dashboard de Vendas/analise_python/img/total_vendas_regiao.png") 

# 👀 Exibe o gráfico
plt.tight_layout()
plt.show()
```

## 💡 Insights: Total de Vendas por Região

📍 **Sudeste** lidera com folga o volume de vendas, representando a maior fatia do total.  
Isso pode indicar uma concentração maior de clientes ou uma demanda mais aquecida na região.

📍 **Sul** e **Centro-Oeste** aparecem como mercados relevantes, embora com valores significativamente menores que o Sudeste.

📉 **Norte** e **Nordeste** apresentam os menores volumes de vendas.  
Pode ser interessante investigar se há menor cobertura de atuação nessas regiões ou outras barreiras de acesso.

---

```python
# 📍 Valor total vendido por região
vendas_regiao = df.groupby('Região')['Valor Total'].sum().sort_values(ascending=False)

# 🧁 Gráfico: Distribuição de Vendas por Região
plt.figure(figsize=(8, 8))
plt.pie(vendas_regiao, labels=vendas_regiao.index, autopct='%1.1f%%', colors=sns.color_palette('pastel'), startangle=90)
plt.title("Distribuição de Vendas por Região")
plt.axis('equal')

# 💾 Salva o gráfico
plt.savefig("C:/Users/Aline.Aline/Desktop/GitHub/Projeto 1 - Dashboard de Vendas/analise_python/img/distribuicao_vendas_regiao.png") 

# 👀 Exibe o gráfico
plt.tight_layout()
plt.show()
```
